```python
import pandas as pd
df=pd.read_csv('result.csv')
```

```python
df.drop(df.index[[15,18,76,83,234,242,373,377,532,536,612,614,713,715,717,783,810,814,819]], inplace=True)
```

```python
df.columns = ['ROLLNO','NAME','College Name','Semester','Degree','subject1','subject2',
              'subject3','subject4','subject5','subject6','subject7','subject8','subject9',
              'subject10','subject11','subject12','subject13','subject14','subject15','subject16','subject17','subject18','SGPA'	]
```

```python
df['SGPA']=df['SGPA'].astype(float)
x=df['SGPA'].mean()
df['SGPA'].fillna(x,inplace=True)
df.fillna('0',inplace=True)
```

```python
g=df.groupby('Degree')
#for degree,degree_df in g:
 # print(degree)
  #print(degree_df)
```

**College Result Analysis**

```python
import matplotlib.pyplot as plt
import numpy as np
topper=df.nlargest(3,columns=['SGPA'],keep='first')
print("Top three students:\n",topper[['NAME','SGPA','Degree']],end="\n\n")
print("Number of Students passed:",sum(df['SGPA'] >= 5))
print("Number of Students failed:",sum(df['SGPA'] < 5),end="\n\n")
print("Students scoring 10 SGPA:",((df['SGPA']==10)).sum())
print("Students scoring between 9 SGPA and 10 SGPA:",((df['SGPA'] >= 9)&(df['SGPA']<10)).sum())
print("Students scoring between 8 SGPA to 9 SGPA:",((df['SGPA'] >= 8)&(df['SGPA']<9)).sum())
print("Students scoring between 7 SGPA to 8 SGPA:",((df['SGPA'] >= 7)&(df['SGPA']<8)).sum())
print("Students scoring between 6 SGPA to 7 SGPA:",((df['SGPA'] >= 6)&(df['SGPA']<7)).sum())
print("Students scoring between 5 SGPA to 4 SGPA:",((df['SGPA'] >= 5)&(df['SGPA']<6)).sum())
print("Students scoring less than 5 SGPA:",sum(df['SGPA'] <5),end="\n\n")

y = np.array([((df['SGPA']==10)).sum(), ((df['SGPA'] >= 9)&(df['SGPA']<10)).sum(),
              ((df['SGPA'] >= 8)&(df['SGPA']<9)).sum(),((df['SGPA'] >= 7)&(df['SGPA']<8)).sum(),
              ((df['SGPA'] >= 6)&(df['SGPA']<7)).sum(),((df['SGPA'] >= 5)&(df['SGPA']<6)).sum(),sum(df['SGPA'] <5)])
x = np.array(["less than 5","5-6","6-7","7-8","8-9","9-10","10"])
plt.xlabel("SGPA of student")
plt.ylabel("Number of students")
plt.bar(x, y, width = 0.5)
plt.show()
```
